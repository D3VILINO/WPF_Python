## Einsteigeraufgaben
### 1. Einfaches Liniendiagramm
Zeichne den Funktionsgraphen y = x^2 für x∈[−10,10].

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10)
y = x**2

fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

### 2. Mehrere Linien in einem Diagramm
Zeichne y = x, y = x^2 und y = x^3 im selben Plot mit Legende.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 200, endpoint=True)
f1 = x**2
f2 = x**3

plt.axis([-10, 10, -10, 20])
plt.plot(x, f1, label="x^2")
plt.plot(x, f2, label="x^3")

plt.legend()
plt.show()
```

### Achsenbeschriftung und Titel

Füge Titel, x-Achsen- und y-Achsen-Beschriftung zu einem Diagramm hinzu.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 200, endpoint=True)
f1 = x**2
f2 = x**3

plt.axis([-10, 10, -10, 20])
plt.title("Titel")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.plot(x, f1, label="x^2")
plt.plot(x, f2, label="x^3")
plt.legend()
plt.show()
```

### Linienstil und Farben
Zeichne eine Linie mit gestrichelter Linie, roter Farbe und Punktmarkierungen.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 5, endpoint=True)
f1 = x

plt.axis([-10, 10, -10, 10])
plt.title("Titel")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.plot(x, f1, "--ro", label="y=x",)
plt.show()
```

## Fortgeschrittene Aufgaben
### Subplots verwenden 
Erstelle ein Diagramm mit 2x2 Subplots, die verschiedene Funktionen zeigen.

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axs = plt.subplots(2, 2)

x = np.linspace(-5, 5, endpoint=True)
f1 = x
f2 = x**2
f3 = x**3
f4 = x**4

axs[0,0].plot(x, f1)
axs[0,1].plot(x, f2)
axs[1,0].plot(x, f3)
axs[1,1].plot(x, f4)

plt.show()
```

### Balkendiagramm
Visualisiere die Umsätze von 5 Produkten als Balkendiagramm.

### Histogramm
Erstelle ein Histogramm von 1000 Zufallszahlen aus einer Normalverteilung

```python
import random
import matplotlib.pyplot as plt

random_values = []
for _ in range(1000):
  random_values.append(random.gauss(mu=50, sigma=0.6))

plt.hist(random_values, bins=70)
plt.show()
```

### Scatterplot
Erstelle ein Streudiagramm mit Zufallsdaten und färbe die Punkte abhängig von
einem dritten Merkmal ein (z. B. Größe, Farbe).

```python
import random
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(1, 100, 100, endpoint=True)
y_change = [random.uniform(-8, 8) for i in range(0,100)]

f1 = x
colors = ['red' if abs(y) >= 4 else 'blue' for y in y_change]

plt.scatter(x, f1 + y_change, color=colors)
plt.show()
```

### Fehlerbalken
Erstelle ein Liniendiagramm mit zufälligen Daten und zeige Fehlerbalken an.

(Mit ChatGPT erstellt, weil ich absolut keine Ahnung hatte, wie ich es erstellen soll und was es ist. Ich habe es eingefügt um eine völlständige Dokumentation für mich zu haben)

```python
import matplotlib.pyplot as plt
import numpy as np

# Erzeuge x-Werte (z. B. Zeit oder Index)
x = np.linspace(0, 10, 20)

# Erzeuge zufällige y-Werte (z. B. Messungen)
y = np.random.normal(loc=5, scale=1, size=20)

# Erzeuge zufällige Fehlerwerte (z. B. Standardabweichung)
y_err = np.random.uniform(0.2, 0.5, size=20)

# Zeichne das Liniendiagramm mit Fehlerbalken
plt.errorbar(x, y, yerr=y_err, fmt='-o', ecolor='red', capsize=5, label='Messwerte mit Fehler')

# Achsenbeschriftung und Legende
plt.xlabel("x-Achse")
plt.ylabel("y-Achse")
plt.title("Liniendiagramm mit Fehlerbalken")
plt.legend()

# Zeige das Diagramm
plt.show()
```

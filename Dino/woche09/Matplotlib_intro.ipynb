## [Matplotlib](https://matplotlib.org/) - Visualisierung mit Python

`Matplotlib` ist eine Python-Bibliothek zur Erstellung von 2D-Plots und Diagrammen. Es bietet eine umfangreiche Palette an Funktionen und Anpassungsmöglichkeiten für die Visualisierung von Daten und unterstützt verschiedene Ausgabeformate wie PNG, PDF und SVG. Matplotlib wird häufig in der Datenanalyse und -visualisierung, der wissenschaftlichen Forschung und der Erstellung von Präsentationen eingesetzt. 

```python
import random
import matplotlib.pyplot as plt

random_values = []
for _ in range(100000):
    random_values.append(random.gauss(mu=100, sigma=20))

print(random_values[:10])

plt.hist(random_values, bins=500, alpha=0.5)
plt.ylabel("count")
plt.grid(True)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2 * np.pi, 200)
y = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
labels = 'FI-A 13', 'FI-A 14' , 'FI-A 15'
sizes = [4, 10, 4]

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels)
```
